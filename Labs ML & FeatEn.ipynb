## Categorical Variables

Here we will try to reduce the number of categories. An ideal way would have been to group the states into 
regions. But in this case we will group all the states with counts less than 2500 into one category "other"

```python
df = pd.DataFrame(categorical['STATE'].value_counts()).reset_index()

df.columns = ['state', 'count']
other_states = list(df[df['count']<2500]['state'])

def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x
    
categorical['STATE'] = list(map(clean_state, categorical['STATE']))
```

For the column NEXTDATE, seeing the descriptions, I have decided to fill the empty values with the sum of FISTDATE and TIMELAG
>FISTDATE Date of first gift
>
>NEXTDATE Date of second gift  
>
>TIMELAG Number of months between first and second gift

```python
for i in range(len(numerical['NEXTDATE'])):
    if numerical['NEXTDATE'][i] > 0: 
        numerical['NEXTDATE'][i] = numerical['NEXTDATE'][i]
    else:
        numerical['NEXTDATE'][i] = numerical['FISTDATE'][i] + numerical['TIMELAG'][i]

        
```
